# Gaussian Processes (GP)

Used to infer distribution over the parameters of a parametric function or a distribution over the label given test points and parameters

A GP is a random process where any point $x \in \mathbb{R^d}$ is assigned a random variable $f(\mathbf{x})$ and where the joint distribution of a finite number of these variables $p(f(\mathbf{x_1},\cdots,f(\mathbf{x_N}))$ is itself Gaussian: 

\begin{align}
p(f \mid \mathbf{X}) \sim \mathcal{N}(f\mid\mu,\mathbf{K})
\end{align}

$\mathbf{K}$ is a positive definite kernel function or covariance function, which means a GP is a distribution over functions whose shape (smoothness, skew, etc. ) is defined by this kernel. Typically the Gaussian, or squared exponential kernel, or RBF kernel is used. 

A GP prior $p(f\mid\mathbf{X})$ can be converted into a GP posterior $p(f\mid\mathbf{X},\mathbf{y})$ after having observed some data $\mathbf{y}$. The posterior can then be used to make predictions $f_*$ given new input $\mathbf{X_*}$:

\begin{align}
p(f_* \mid \mathbf{X_*},\mathbf{X},\mathbf{y}) &= \int p(f_* \mid \mathbf{X_*},\mathbf{f}) df \\
&= \mathcal{N}(f_*\mid\mu_*,\mathbf{\Sigma_*})
\end{align}

Eq (2-3) is the posterior predictive distribution which is also a Gaussian with mean $\mu_*$ and covariance $\Sigma_*$. By definition of the GP, the joint distribution of observed data $\mathbf{y}$ and predictions $f_*$ is:

# Implementation

```python
# functions for plotting

def plot_gp(mu, cov, X, X_train=None, Y_train=None, samples=[]):
    X = X.ravel()
    mu = mu.ravel()
    uncertainty = 1.96 * np.sqrt(np.diag(cov))
    
    fig = plt.figure(figsize=(10,7))
    plt.fill_between(X, mu + uncertainty, mu - uncertainty, alpha=0.25, color='red')
    plt.plot(X, mu, label='Mean')
    for i, sample in enumerate(samples):
        plt.plot(X, sample, lw=1, ls='--', label=f'Sample {i+1}')
    if X_train is not None:
        plt.plot(X_train, Y_train, 'rx')
    plt.legend()

def plot_gp_2D(gx, gy, mu, X_train, Y_train, title, i):
    ax = plt.gcf().add_subplot(1, 2, i, projection='3d')
    ax.plot_surface(gx, gy, mu.reshape(gx.shape), cmap=cm.coolwarm, linewidth=0, alpha=0.2, antialiased=False)
    ax.scatter(X_train[:,0], X_train[:,1], Y_train, c=Y_train, cmap=cm.coolwarm)
    ax.set_title(title)
```

```python
# kernel function

def kernel(X1, X2, l=1.0, sigma_f=1.0):
    '''
    Isotropic squared exponential kernel. Computes 
    a covariance matrix from points in X1 and X2.
        
    Args:
        X1: Array of m points (m x d).
        X2: Array of n points (n x d).

    Returns:
        Covariance matrix (m x n).
    '''
    sqdist = np.sum(X1**2, 1).reshape(-1, 1) + np.sum(X2**2, 1) - 2 * np.dot(X1, X2.T)
    return sigma_f**2 * np.exp(-0.5 / l**2 * sqdist)
```

Define a prior over functions with a mean of zero and a covariance computed with $l=1$ and $\sigma_f=1$. Here, we draw three random samples and plot with the zero mean and 95% confidence interval (computed from the diagonal of the cov matrix).

```python
# generate a finite number of points
X = np.arange(-5,5, 0.2)
X = X.reshape(-1,1) # row to column

# mean and covariance of the prior
mu = np.zeros(X.shape)
cov = kernel(X,X)

# draw three samples from the prior
samples = np.random.multivariate_normal(mu.ravel(), cov, 3)

# plot GP mean, confidence interval, and the three samples
plot_gp(mu, cov, X, samples=samples)
```

## Prediction from noise-free training data

To compute mean and covariance of the posterior distribution we implement equations 5 and 6

```python
def posterior_predictive(X_s, X_train, Y_train, l=1.0, sigma_f=1.0, sigma_y=1e-8):
    '''  
    Computes the suffifient statistics of the GP posterior predictive distribution 
    from m training data X_train and Y_train and n new inputs X_s.
    
    Args:
        X_s: New input locations (n x d).
        X_train: Training locations (m x d).
        Y_train: Training targets (m x 1).
        l: Kernel length parameter.
        sigma_f: Kernel vertical variation parameter.
        sigma_y: Noise parameter.
    
    Returns:
        Posterior mean vector (n x d) and covariance matrix (n x n).
    '''
    K = kernel(X_train, X_train, l, sigma_f) + sigma_y**2 * np.eye(len(X_train))
    K_s = kernel(X_train, X_s, l, sigma_f)
    K_ss = kernel(X_s, X_s, l, sigma_f) + 1e-8 * np.eye(len(X_s))
    K_inv = inv(K)
    
    # Equation (5)
    mu_s = K_s.T.dot(K_inv).dot(Y_train)

    # Equation (6)
    cov_s = K_ss - K_s.T.dot(K_inv).dot(K_s)
    
    return mu_s, cov_s
```

Now we apply them to the noise-free training data `X_train` and `Y_train`. We draw three samples from the posterior predictive distribution and plot them with the mean, confidence interval and training data.

```python
# noise-free training data
X_train = np.array([-4, -3, -2, -1, 1]).reshape(-1, 1)
Y_train = np.sin(X_train)

# compute mean and covariance of the posterior predictive distribution
mu_s, cov_s = posterior_predictive(X, X_train, Y_train)

# draw three samples
samples = np.random.multivariate_normal(mu_s.ravel(), cov_s, 3)

# plot
plot_gp(mu_s, cov_s, X, X_train=X_train, Y_train=Y_train, samples=samples)
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x) + np.random.random(100) * 0.8

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

f = plt.figure(figsize=(8,5))
ax = f.add_subplot(1,1,1)

plt.plot(x, smooth(y,1), lw=2, c='C0')
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
plt.savefig('images_for_pres/noisy.png',dpi=300)
plt.show()
```
