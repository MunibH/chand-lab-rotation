```python
import scipy.io
import pandas as pd
from elephant.gpfa import GPFA
import numpy as np
import quantities as pq
import neo
import pickle
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

load gpfa data

```python
with open('data/spikeTimes.pickle', 'rb') as filename:
    st = pickle.load(filename)
    
behav = pd.read_csv('data/RT_cue_choice.csv',header=None, names=['RT','cue','choice'])
```

```python
# add a movement onset column
behav['mvmt'] = behav['RT'] + 400
```

```python
behav.head()
```

```python
numTrials = len(st)
numNeurons = len(st[0])
```

Raster Plot

```python
fig = plt.figure(figsize=(10,6))
for i, spiketrain in enumerate(st[0]):
    plt.plot(spiketrain, np.ones_like(spiketrain)*i, ls='', marker='|')

plt.title('Raster plot of trial 1')
plt.xlabel('time (ms)')
plt.ylabel('Neuron')
plt.show()
```

Apply GPFA (fit with first 20 trials for computational ease)

```python
# use a small number of trials for now
# st = st[:numTrials//20]
# numTrials = len(st)
```

```python
# specify fitting parameters
bin_size = 20 * pq.ms
latent_dimensionality = 2

gpfa_2dim = GPFA(bin_size=bin_size, x_dim=latent_dimensionality)
```

```python
# gpfa_2dim.fit(st[:numTrials//2])
# print(gpfa_2dim.params_estimated.keys())
```

```python
# trajectories = gpfa_2dim.transform(st[numTrials//2:])
```

```python
# here we just reuse the existing instance of the GPFA() class as we use the same fitting parameters as before

trajectories_all = gpfa_2dim.fit_transform(st)
```

```python
# plot until mvmt onset

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(25, 15))

linewidth_single_trial = 0.5
color_single_trial = 'C0'
alpha_single_trial = 0.5

linewidth_trial_average = 2
color_trial_average = 'C1'

########################################################################

ax1.set_title('Latent dynamics extracted by GPFA')
ax1.set_xlabel('Dim 1')
ax1.set_ylabel('Dim 2')
ax1.set_aspect(1)
trial_idx = 0
for single_trial_trajectory in trajectories_all:

    ax1.plot(single_trial_trajectory[0], single_trial_trajectory[1], '-', lw=linewidth_single_trial, c=color_single_trial, alpha=alpha_single_trial)
    trial_idx += 1
    
average_trajectory = np.mean(trajectories_all, axis=0)
ax1.plot(average_trajectory[0], average_trajectory[1], '-', lw=linewidth_trial_average, c=color_trial_average, label='Trial averaged trajectory')
ax1.legend()

########################################################################

trial_to_plot = 0
ax2.set_title(f'Trajectory for trial {trial_to_plot}')
ax2.set_xlabel('Time [ms]')
times_trajectory = np.arange(len(trajectories_all[trial_to_plot][0])) * bin_size.rescale('ms')

# get mvmt onset index
m = behav['mvmt'][trial_to_plot]
for i in range(20):
    if m % 20 == 0:
        break
    m += 1
t_idx = np.where(times_trajectory==m)[0][0]

ax2.plot(times_trajectory[:t_idx], trajectories_all[0][0][:t_idx], c='C0', label="Dim 1, fitting with all trials")
ax2.plot(times_trajectory[:t_idx], trajectories_all[0][1][:t_idx], c='C1', label="Dim 2, fitting with all trials")
ax2.legend()

########################################################################

# get fast and slow trial idx
quantiles = behav['RT'].quantile([0.25,0.75])
fastRTs = behav[behav['RT'] <= quantiles[0.25]]
fastRTidx = list(fastRTs.index)
slowRTs = behav[behav['RT'] >= quantiles[0.75]]
slowRTidx = list(slowRTs.index)

ax3.set_title('Latent dynamics of fast RT trials')
ax3.set_xlabel('Dim 1')
ax3.set_ylabel('Dim 2')
ax3.set_aspect(1)
ax4.set_title('Latent dynamics of slow RT trials')
ax4.set_xlabel('Dim 1')
ax4.set_ylabel('Dim 2')
ax4.set_aspect(1)

trial_idx = 0
for single_trial_trajectory in trajectories_all:
    
    if trial_idx in fastRTidx:
        color_single_trial = 'C1'
        alpha_single_trial = 0.9
        
        # get mvmt onset index
        m = behav['mvmt'][trial_idx]
        for i in range(20):
            if m % 20 == 0:
                break
            m += 1
        t_idx = np.where(times_trajectory==m)[0][0]

        ax3.plot(single_trial_trajectory[0][:t_idx], single_trial_trajectory[1][:t_idx], '-', lw=linewidth_single_trial, c=color_single_trial, alpha=alpha_single_trial)
        trial_idx += 1
        
    elif trial_idx in slowRTidx:
        color_single_trial = 'C0'
        alpha_single_trial = 0.4
        
        # get mvmt onset index
        m = behav['mvmt'][trial_idx]
        for i in range(20):
            if m % 20 == 0:
                break
            m += 1
        t_idx = np.where(times_trajectory==m)[0][0]

        ax4.plot(single_trial_trajectory[0][:t_idx], single_trial_trajectory[1][:t_idx], '-', lw=linewidth_single_trial, c=color_single_trial, alpha=alpha_single_trial)
        trial_idx += 1
        
    else:
        trial_idx += 1
        continue
    
#     # get mvmt onset index
#     m = behav['mvmt'][trial_idx]
#     for i in range(20):
#         if m % 20 == 0:
#             break
#         m += 1
#     t_idx = np.where(times_trajectory==m)[0][0]
    
    
    
#     ax3.plot(single_trial_trajectory[0][:t_idx], single_trial_trajectory[1][:t_idx], '-', lw=linewidth_single_trial, c=color_single_trial, alpha=alpha_single_trial)
#     trial_idx += 1
#     if trial_idx == 100:
#         break


plt.show()
```

```python
# bin the trials by RT

def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3

nbins = 20 # actually 19
quantile_array = np.arange(1/nbins,1,1/nbins)
quantiles = behav['RT'].quantile(quantile_array)
trials_per_bin = {}
count = 1
for i in quantiles.index:
    if count==1:
        temp = behav[behav['RT'] <= quantiles[i]]
        trials_per_bin[i] = list(temp.index)
    elif count==len(quantiles):
        temp = behav[behav['RT'] >= quantiles[i]]
        trials_per_bin[i] = list(temp.index)
    else:
        t1 = list(behav[behav['RT'] >= quantiles[0.05]].index)
        t2 = list(behav[behav['RT'] <= quantiles[0.1]].index)
        trials_per_bin[i] = intersection(t1,t2)
    
    count += 1

trials_per_bin.keys()
```

```python
# average trials per bin
avg_traj = {}
for i in quantiles.index:
    avg_traj[i] = np.mean(trajectories_all[trials_per_bin[i]], axis=0)
    
```
